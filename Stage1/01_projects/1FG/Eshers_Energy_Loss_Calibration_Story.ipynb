## Calibrating Beam Loss Monitor Measurements
Beam Loss Monitors do not give a quantiative measure of energy lost within the synchrotron. The nature of these monitors is to show the regions of the synchrotron ring where protons may be lost. It can sometimes be the case that detected protons can rejoin the ring and ultimately are not lost. Ultimately, beam loss monitors alone cannot give a complete picture of the energy loss in the synchrotron.  
There must be a calibration of the beam loss monitors, perhaps via comparison with the intensity monitors, which can measure energy in terms of loss of beam intensity.  
Over several decades, such a calibration has been atttempted, the results of which are detailed below.



## Chris' Method
*Estimates of Energy Distribution of Lost Protons in the ISIS Synchrotron, 
C M Warsop, 21/03/2000* 

Beam loss is assessed via _**intensity toroid measurements**_ and **_beam loss monitor measurements._**  

Toroids provide calibrated measurements of circulating protons to an accuracy of about 0.5%, their sensitivity is limited to about $0.5 ⋅ 10^{11}$ ppp (protons per pulse). For large losses, i.e. the first ms or so of the cycle, these provide good measurements.

After the first few ms losses quickly fall to levels below $0.5 ⋅ 10^{11}$ ppp, but are still important because of the higher activation levels associated with higher energy particles. 

The intensity monitor signal, $I[t]$, gives a voltage proportional to the circulating number of protons, $N[t]$, at a given time t in the cycle.

$$ N[t]=K_I⋅I[t] $$
where $K_I=4.0 ⋅ 10^{12}$ protons per volt.  
It is unclear where this constant of proportionality was sourced from and it should be noted that this value is constant and therefore independent of time.
  
The loss rate, in terms of protons per second, is:  

$$ \frac{dN[t]}{dt}=K_I⋅\frac{dI[t]}{dt}$$  


For a digitised signal, sampled it times $t_i$, at intervals of $t_s$, the above equation becomes the numerical differentiation of the signal:  

$$\frac{dN[t_i]}{dt}≈K_I⋅\frac{I[t_i]-I[t_{i-1}]}{t_s} $$

The total number of protons lost over a time interval is given by:  

$$ΔN=∫_{t_1}^{t_2}\frac{dN[t]}{dt}⋅dt=K_I⋅∫_{t_1}^{t_2}\frac{dI[t]}{dt}⋅dt=K_I (I[t_2 ]-I[t_1 ]) $$

Similarly, the numerical integration, via the trapezium rule, of the above equation gives:

$$ΔN=∫_{t_1}^{t_2}\frac{dN[t]}{dt}⋅dt≈K_I⋅∑_{t_i=t_1}^{t_2}\biggr(\frac{I[t_i]-I[t_{i-1} ]}{t_s}\biggr) ⋅t_s≈K_I (I[t_2]-I[t_1])$$

In this way, through the manipulation of the intensity monitor signals, we are able to obtain an estimation of the number of protons lost over selected time intervals.

### Beam Loss Monitor Sum Signal Measurements
Since the intensity toroid monitor signals are limited to a sensitivity of $0.5 ⋅ 10^{11}$ ppp, and after the first few ms losses quickly fall to below $0.5 ⋅ 10^{11}$ ppp, we are unable to continue to assess beam loss using intensity toroid measurements. We must estimate these losses using beam loss monitors instead. 

As previously discussed, the beam loss monitors do not provide accurate quantative data in regards to the number of protons lost, rather they provide a more qualitative insight into the sections of the ring superperiods where escaped protons have been detected. 

The beam loss monitor sum signal provides a voltage which is proportional to the total number of protons lost during the 10 ms pulse period.

It is assumed that, for a given energy, the beam loss sum signal $BL_s[t]$ (in volts) is proportional to the total beam loss $\frac{dN}{dt}$:

$$\frac{dN[t]}{dt}=K_{BL}[t]⋅BL_s[t]$$

where $K_{BL}[t]$ is a time (energy) dependent calibration in protons per volt second.

For a digitised signal, sampled it times $t_i$, at intervals of $t_s$:
$$ΔN=∫_{t_1}^{t_2}\frac{dN[t]}{dt}⋅dt≈\sum_{t_i=t_1}^{t_2} K_{BL} [t_i]⋅BL_s [t_i] ⋅t_s$$

Through the analysis of the intensity monitor signal and the integration of the summed beam loss, we aim to find a calibration curve which is essentially the time (energy) dependent function of the BL calibration coefficient, $K_{BL}$. 

### Integrating Beam Loss
One method of calibrating the BLMs in terms of energy compares the integrated sum beam loss against the change in protons measured by the intensity monitors.
Integrating the sum beam loss is done via the trapezium rule.
We can break the sum signal into arbitrary time intervals, and calculate the area of the trapezium enclosed by the time interval selected.

## Hayley and Bryan's investigation
Hayley and Bryan set about conducting an experiment to plot BLM response as a function of beam energy. 
They went about this by intentionally creating beam loss and measuring the loss and reduction in beam intensity, using a diluted chopped beam.
The horizontal orbit was constant throughout the cycle. 
- Beam was lost at 1-5 ms by creating an orbital bump at R1HM1
- Beam was lost from 5-10 ms by creating an orbital bump on top of a radial loop change.
